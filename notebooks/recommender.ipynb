```python
import pandas as pd

# Load data
scheme = pd.read_csv("mutual_fund_analytics/data/raw/07_scheme_performance.csv")

# Clean text columns
scheme["risk_grade"] = scheme["risk_grade"].astype(str).str.strip()
scheme["sharpe_ratio"] = pd.to_numeric(scheme["sharpe_ratio"], errors="coerce")

# User input
risk = input("Enter Risk Appetite (Low/Moderate/High): ").strip().lower()

# Filter based on risk appetite
if risk == "low":
    filtered = scheme[scheme["risk_grade"] == "Low"]

elif risk == "moderate":
    filtered = scheme[scheme["risk_grade"].isin(["Moderate", "Moderately High"])]

elif risk == "high":
    filtered = scheme[scheme["risk_grade"].isin(["High", "Very High"])]

else:
    print("Invalid Risk Appetite")
    exit()

# Sort by Sharpe Ratio
recommendation = (
    filtered.sort_values(by="sharpe_ratio", ascending=False)
            .head(3)
)

print("\nTop 3 Recommended Funds\n")

print(
    recommendation[
        [
            "scheme_name",
            "fund_house",
            "category",
            "risk_grade",
            "sharpe_ratio"
        ]
    ]
)
```
