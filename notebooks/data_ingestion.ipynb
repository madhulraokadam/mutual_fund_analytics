```python
import pandas as pd

fund_master = pd.read_csv(r"C:\Users\madhu\mutual_fund_analytics\data\raw\01_fund_master.csv")
nav_history = pd.read_csv(r"C:\Users\madhu\mutual_fund_analytics\data\raw\02_nav_history.csv")
aum_by_fund_house = pd.read_csv(r"C:\Users\madhu\mutual_fund_analytics\data\raw\03_aum_by_fund_house.csv")
monthly_sip_inflows = pd.read_csv(r"C:\Users\madhu\mutual_fund_analytics\data\raw\04_monthly_sip_inflows.csv")
category_inflows = pd.read_csv(r"C:\Users\madhu\mutual_fund_analytics\data\raw\05_category_inflows.csv")
industry_folio_count = pd.read_csv(r"C:\Users\madhu\mutual_fund_analytics\data\raw\06_industry_folio_count.csv")
scheme_performance = pd.read_csv(r"C:\Users\madhu\mutual_fund_analytics\data\raw\07_scheme_performance.csv")
investor_transactions = pd.read_csv(r"C:\Users\madhu\mutual_fund_analytics\data\raw\08_investor_transactions.csv")
portfolio_holdings = pd.read_csv(r"C:\Users\madhu\mutual_fund_analytics\data\raw\09_portfolio_holdings.csv")
benchmark_indices = pd.read_csv(r"C:\Users\madhu\mutual_fund_analytics\data\raw\10_benchmark_indices.csv")

datasets = {
    "01_fund_master": fund_master,
    "02_nav_history": nav_history,
    "03_aum_by_fund_house": aum_by_fund_house,
    "04_monthly_sip_inflows": monthly_sip_inflows,
    "05_category_inflows": category_inflows,
    "06_industry_folio_count": industry_folio_count,
    "07_scheme_performance": scheme_performance,
    "08_investor_transactions": investor_transactions,
    "09_portfolio_holdings": portfolio_holdings,
    "10_benchmark_indices": benchmark_indices
}

for name, df in datasets.items():

    print("\n" + "=" * 80)
    print("DATASET:", name)

    print("\nSHAPE:")
    print(df.shape)

    print("\nDATA TYPES:")
    print(df.dtypes)

    print("\nFIRST 5 ROWS:")
    print(df.head())

    print("\nMISSING VALUES:")
    print(df.isnull().sum())

    print("\nDUPLICATE ROWS:")
    print(df.duplicated().sum())
```

```python
import requests
import pandas as pd

schemes = {
    "SBI_Bluechip": 119551,
    "ICICI_Bluechip": 120503,
    "Nippon_LargeCap": 118632,
    "Axis_Bluechip": 119092,
    "Kotak_Bluechip": 120841
}

for name, code in schemes.items():

    print(f"\nChecking {name} ({code})")

    url = f"https://api.mfapi.in/mf/{code}"

    response = requests.get(url)

    print("Status Code:", response.status_code)

    try:
        data = response.json()

        df = pd.DataFrame(data["data"])

        df.to_csv(
            rf"C:\Users\madhu\mutual_fund_analytics\data\raw\{name}.csv",
            index=False
        )

        print(f"{name} saved successfully")

    except Exception as e:
        print(f"Error for {name}:")
        print(e)
        print("Response Text:")
        print(response.text[:500])
```
